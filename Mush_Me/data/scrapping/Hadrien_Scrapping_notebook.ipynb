```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
url = 'https://www.guidedeschampignons.com/tous-les-champignons/'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

html = soup.find_all('a', class_="woocommerce-LoopProduct-link woocommerce-loop-product__link")

html
```

```python
mushrooms_FR = []

for link in html:
    mushroom = {}
    
    mushroom['name'] = link.find('h2',class_='woocommerce-loop-product__title').string.capitalize()
    
    mushroom['link'] = link.get('href')
    
    mushrooms_FR.append(mushroom)

mushrooms_FR
```

```python
mushrooms_FR
```

```python
mushrooms_df_FR = pd.DataFrame.from_dict(mushrooms_FR)
mushrooms_df_FR
```

```python
mushrooms_df_FR.to_csv('mushrooms_FR.csv')
```

```python
url_2 = 'https://www.mushroom.world/mushrooms/namelist'

response_2 = requests.get(url_2)

soup_2 = BeautifulSoup(response_2.content, 'html.parser')

html_2 = soup_2.find_all('div', class_='item')

html_2
```

```python
mushrooms_2 = []

for link in html_2:
    mushroom = {}
    
    mushroom['Name'] = link.find('a').string
    
    mushroom['Link'] = re.findall('(?:(?:https?|ftp):\/\/)?[\w/\-?=%.]+\.[\w/\-&?=%.]+', str(link))[0]
    
    mushrooms_2.append(mushroom)

mushrooms_2
```

```python
for mushroom in mushrooms_2:
    url = mushroom['Link']
    
    response_3 = requests.get(url)
    
    soup_3 = BeautifulSoup(response_3.content, 'html.parser')
    
    html_3 = soup_3.find('div', class_='container body-content')
    
    mushroom['Edibility'] = html_3.find_all('div',class_='textus')[3].string
    
    mushroom['Family'] = html_3.find_all('div',class_='textus')[0].string
    
    mushroom['Location'] = html_3.find_all('div',class_='textus')[1].string
    
    mushroom['Dimensions'] = html_3.find_all('div',class_='textus')[2].string
    
mushrooms_2
```

```python
mushrooms_2_df = pd.DataFrame.from_dict(mushrooms_2)
mushrooms_2_df
```

```python
mushrooms_2_df.to_csv('mushrooms_edibility.csv')
```
